```python
def change_month(x):
    month = df["Reported Month"]
    for y in month:
        if (y == 'Mar'):
            x = x.replace(y, 'Mars')
        if (y == 'Feb'):
            x = x.replace(y, 'Février')
        if (y == 'Jan'):
            x = x.replace(y, 'Janvier')
    return x
```
